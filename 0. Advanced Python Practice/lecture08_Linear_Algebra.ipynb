```python
# Matrix Additional
ma = [[3,6], [4,5]]
mb = [[5,8], [6,7]]

# Asteriskë¥¼ ì‚¬ìš©í•˜ì—¬ unpackingì‹œì¼œ ê° indexì˜ ìš”ì†Œë¥¼ tupleì—ì„œ ì›ë˜ìƒíƒœë¡œ ë³€í™˜ì‹œí‚´
res = [[sum(row) for row in zip(*t)] for t in zip(ma, mb)]
print(res)
```

```python
# Matrix Scalar-Matrix Product

ma = [[3,6], [4,5]]
alpha = 4
res = [[alpha*element for element in t] for t in ma]
print(res)
```

```python
# Matrix Transpose
ma = [[1,2,3], [4,5,6]]
res = [[element for element in t] for t in zip(*ma)]
print(res)
```

```python
# Matirx Product
ma = [[1,1,2],[2,1,1]]
mb = [[1,1], [2,1], [1,3]]
res = [[sum(a*b for a, b in zip(row_a, col_b)) for col_b in zip(*mb)] for row_a in ma]
print(res)
```
