```python
#*args
def aa(a, *args):
    print(a, args)
    print(type(args))

aa(1,2,3,4,5,6)
```

```python
#**kargs
def rr(a, **kargs):
    print(a, kargs)
    print(type(kargs))

rr(1,2,3,4,5,6)
```

```python
#**kargs
def bb(a, **kargs):
    print(a, kargs)
    print(type(kargs))

bb(1, b=2, c=3, d=4, e=5, f=6)
```

```python
def aa(a, *args):
    print(a, args[0])
    print(type(args))
aa(1, *(2,3,4,5,6))
```

```python
def aa(a, *args):
    print(a, args)
    print(type(args))
aa(1, (2,3,4,5,6))
```

```python
def bb(a, args):
    print(a, *args)
    print(type(args))

bb(1,(2,3,4,5,6))
```

```python
a, b, c = ([1,2], [3,4], [5,6])
print(a,b,c)

data = ([1,2], [3,4], [5,6])
print(*data)
```

```python
def ast(a,b,c,d):
    print(a,b,c,d)

data = {"b":1, "c":2, "d":3}
ast(10, **data)
```

```python
for data in zip(*([1,2], [3,4], [5,6])):
    print(data)
```
