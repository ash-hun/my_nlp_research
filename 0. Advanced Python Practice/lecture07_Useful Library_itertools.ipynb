```python
# module import
import itertools as it
```

```python
# Count()
# Count(start, [step])
# return Infinity Iterator
start = 10
step = 5
it.count(start, step)_#10, 15, 20, 25, ...
```

```python
#Cycle()
#Cycle(p)
# return Infinity Iterator
target = 'ABCDE'
it.cycle(target)_#A B C D E A B C D ...
```

```python
#Repeat()
#Repeat(element, [,n])
# return Infinity Iterator
elem = 10
n = 3
for i in it.repeat(elem, n):
    print(i)
```

```python
# Compress()
# Compress(data, selectors)
# return iterator
# íŠ¹ì • ìš”ì†Œë¥¼ ì¸ë±ìŠ¤ë¡œ í•„í„°ë§
data = 'ABCDEFG'
list(it.compress(data, [1,1,1,0,0,0]))
```

```python
# accumulate()
# accumulate(p, [,func])
# return iterator
# ê° í•­ëª©ì— ëŒ€í•œ ëˆ„ì ê°’
data = [1,2,3,4,5]
list(it.accumulate(data))
```

```python
# filterfalse()
# filterfalse(pred, seq)
# pred(elem)ì´ ê±°ì§“ì¸ seqì˜ ìš”ì†Œë“¤ì„ ë°˜í™˜
list(it.filterfalse(lambda x: x%2, range(10)))
```

$$ {}_n \prod{}_r = n^{r}$$


```python
# Product()
# Product(p,q,[repeat=1])
# ë°ì¹´ë¥´íŠ¸ ê³±(cartesian product), ì¤‘ì²©ëœ for ë£¨í”„ì™€ ë™ì¼í•¨!
target = 'ABC'
t = list(it.product(target, repeat=2))
print(t)
print(len(t))
print('='*100)
Target = ['ABC', '123']
T = list(it.product(*Target))
print(T)
print(len(T))
```

# **Permutations()**

**ìˆœì—´**ì€ ìˆœì„œë¥¼ ìƒê°í•˜ê³  ì¤‘ë³µì„ í—ˆë½í•œì§€ ì•ŠëŠ” ê²½ìš°ë¥¼ ë§í•œë‹¤.

$$ {}_n P {}_r = \frac{n!}{(n-r)!} $$

```python
# Permutations()
# permutations(p, [,r])
# ìˆœì—´
data = "ABCD"
t = list(it.permutations(data, 2))
print(t)
print(len(t))
```

# **Combinations_with_replacement()**

**ì¤‘ë³µì¡°í•©**ì€ ìˆœì„œë¥¼ ìƒê°í•˜ì§€ ì•Šê³  ì¤‘ë³µì„ í—ˆë½í•˜ëŠ” ê²½ìš° ì¤‘ë³µì¡°í•©ì´ë¼ê³  í•œë‹¤.

$$ {}_n H {}_r = {}_{n+r-1} C {}_r$$


```python
# combinations_with_replacement()
# combinations_with_replacement(p, r)
# ì¤‘ë³µì¡°í•©
data = "ABCD"
t = list(it.combinations_with_replacement(data, 2))
print(t)
print(len(t))
```

# **Combinations**
