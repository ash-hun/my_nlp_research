### **ğŸ‘‰ List Comprehension**
- ê¸°ì¡´ list ì‚¬ìš©í•˜ì—¬ ê°„ë‹¨íˆ ë‹¤ë¥¸ listë¥¼ ë§Œë“œëŠ” ê¸°ë²•
- í¬ê´„ì ì¸ list, í¬í•¨ë˜ëŠ” ë¦¬ìŠ¤íŠ¸ë¼ëŠ” ì˜ë¯¸ë¡œ ì‚¬ìš©ë¨
- íŒŒì´ì¬ì—ì„œ ê°€ì¥ ë§ì´ ì‚¬ìš©ë˜ëŠ” ê¸°ë²• ì¤‘ í•˜ë‚˜
- ì¼ë°˜ì ìœ¼ë¡œ for + append ë³´ë‹¤ ì†ë„ê°€ ë¹ ë¦„

```python
res = []
for i in range(10):
    res.append(i)
res
```

```python
res = [i for i in range(10)]
res
```

```python
res = [i for i in range(10) if i%2==0]
res
```

```python
word_1 = "Hello"
word_2 = "World"
res = [i+j for i in word_1 for j in word_2]
res
```

```python
case_1 = ["A", "B", "C"]
case_2 = ["D", "E", "A"]
res = [i+j for i in case_1 for j in case_2]
res
```

```python
res = [i+j for i in case_1 for j in case_2 if not(i==j)] # Filter : iì™€ jê°€ ê°™ë‹¤ë©´ listì— ì¶”ê°€í•˜ì§€ ì•ŠìŒ
res.sort()
res
```

```python
words = "The quick brown fox jumps over the lazy dog".split()
words
```

```python
stuff = [[w.upper(), w.lower(), len(w)] for w in words]
stuff
```

### **ğŸ‘‰ Enumerate**
- Listì˜ elementë¥¼ ì¶”ì¶œí•  ë•Œ ë²ˆí˜¸ë¥¼ ë¶™ì—¬ì„œ ì¶”ì¶œ

```python
# Enumerate
for i, v in enumerate(['tic', 'tac', 'toe']):
    print(i, v)
```

```python
mylist = ["a", "b", "c", "d"]
list(enumerate(mylist))
```

```python
{i:j for i, j in enumerate('Gachon University is an academic institute located in South Korea.'.split())}
```

### **ğŸ‘‰ Zip**
- 2ê°œì˜ listì˜ ê°’ì„ ë³‘ë ¬ì ìœ¼ë¡œ ì¶”ì¶œí•¨

```python
alist = ['a1', 'a2', 'a3']
blist = ['b1', 'b2', 'b3']
for a, b in zip(alist, blist):
    print(a, b)
```

```python
a,b,c = zip((1,2,3), (10,20,30), (100,200,300))
print(a, b, c)
```

```python
[sum(x) for x in zip((1,2,3), (10,20,30), (100,200,300))]
```

```python
alist = ['a1', 'a2', 'a3']
blist = ['b1', 'b2', 'b3']

for i, (a, b) in enumerate(zip(alist, blist)):
    print(i, a, b)
```
