### ğŸ‘‰ **Reduce Function**

- map functionê³¼ ë‹¬ë¦¬ listì— ë˜‘ê°™ì€ í•¨ìˆ˜ë¥¼ ì ìš©í•´ì„œ í†µí•©

```python
from functools import reduce
print(reduce(lambda x, y: x+y, [1,2,3,4,5]))
```

```python
def factorial(n):
    return reduce(lambda x, y: x*y, range(1, n+1))

factorial(5)
```
